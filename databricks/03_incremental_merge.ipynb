```python
incoming_orders = [
    (1002, 102, 502, 1, 750.0, "2026-06-24 09:00:00"),
    (1003, 103, 503, 2, 900.0, "2026-06-24 09:05:00")
]

columns = [
    "order_id",
    "customer_id",
    "product_id",
    "quantity",
    "amount",
    "event_time"
]

incoming_df = spark.createDataFrame(incoming_orders, columns)

display(incoming_df)
```

```python
from delta.tables import DeltaTable

silver_table = DeltaTable.forName(
    spark,
    "ecommerce_project.silver_orders"
)
```

```python
from pyspark.sql.functions import current_timestamp

incoming_df = (
    incoming_df
        .withColumn(
            "processed_timestamp",
            current_timestamp()
        )
)
```

```python
(
    silver_table.alias("target")
    .merge(
        incoming_df.alias("source"),
        "target.order_id = source.order_id"
    )
    .whenMatchedUpdateAll()
    .whenNotMatchedInsertAll()
    .execute()
)
```

Pipeline Metrics

```python
bronze_count = spark.table("ecommerce_project.bronze_orders").count()
```

```python
silver_count = spark.table("ecommerce_project.silver_orders").count()
```

```python
rejected_count = bronze_count - silver_count
```

```python
from pyspark.sql.functions import current_timestamp

metrics = [
    (
        bronze_count,
        silver_count,
        rejected_count
    )
]

metrics_df = spark.createDataFrame(
    metrics,
    [
        "bronze_count",
        "silver_count",
        "rejected_count"
    ]
).withColumn(
    "pipeline_run",
    current_timestamp()
)

display(metrics_df)
```

```python
metrics_df.write \
    .format("delta") \
        .mode("append") \
            .saveAsTable("ecommerce_project.pipeline_metrics")
```
